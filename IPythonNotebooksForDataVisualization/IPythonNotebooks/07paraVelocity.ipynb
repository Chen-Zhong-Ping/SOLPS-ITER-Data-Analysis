some preparation calculation

calculate carbon velocity

```python
# first calculate total carbon ion density for later use
SD_nCarbonIon = SD_nIon3.IyIx + SD_nIon4.IyIx + SD_nIon5.IyIx + SD_nIon6.IyIx + SD_nIon7.IyIx + SD_nIon8.IyIx

FS_nCarbonIon = FS_nIon3.IyIx + FS_nIon4.IyIx + FS_nIon5.IyIx + FS_nIon6.IyIx + FS_nIon7.IyIx + FS_nIon8.IyIx

XD_nCarbonIon = XD_nIon3.IyIx + XD_nIon4.IyIx + XD_nIon5.IyIx + XD_nIon6.IyIx + XD_nIon7.IyIx + XD_nIon8.IyIx


# then calculate the average carbon ion velocity
# take out the outer divertor part

SD_uParallelCarbon_outerdiv = numpy.transpose(((SD_uIon3.IyIx * SD_nIon3.IyIx
                                              + SD_uIon4.IyIx * SD_nIon4.IyIx
                                              + SD_uIon5.IyIx * SD_nIon5.IyIx
                                              + SD_uIon6.IyIx * SD_nIon6.IyIx
                                              + SD_uIon7.IyIx * SD_nIon7.IyIx
                                              + SD_uIon8.IyIx * SD_nIon8.IyIx)/SD_nCarbonIon)[:, SD0203_mesh.ixcut[1]:])

FS_uParallelCarbon_outerdiv = numpy.transpose(((FS_uIon3.IyIx * FS_nIon3.IyIx
                                              + FS_uIon4.IyIx * FS_nIon4.IyIx
                                              + FS_uIon5.IyIx * FS_nIon5.IyIx
                                              + FS_uIon6.IyIx * FS_nIon6.IyIx
                                              + FS_uIon7.IyIx * FS_nIon7.IyIx
                                              + FS_uIon8.IyIx * FS_nIon8.IyIx)/FS_nCarbonIon)[:, FS0308_mesh.ixcut[1]:])

XD_uParallelCarbon_outerdiv = numpy.transpose(((XD_uIon3.IyIx * XD_nIon3.IyIx
                                              + XD_uIon4.IyIx * XD_nIon4.IyIx
                                              + XD_uIon5.IyIx * XD_nIon5.IyIx
                                              + XD_uIon6.IyIx * XD_nIon6.IyIx
                                              + XD_uIon7.IyIx * XD_nIon7.IyIx
                                              + XD_uIon8.IyIx * XD_nIon8.IyIx)/XD_nCarbonIon)[:, XD0130_mesh.ixcut[1]:])
```

compare carbon velocity

```python
print(SD_uParallelCarbon_outerdiv.max())
print(FS_uParallelCarbon_outerdiv.max())
print(XD_uParallelCarbon_outerdiv.max())

print(SD_uParallelCarbon_outerdiv.min())
print(FS_uParallelCarbon_outerdiv.min())
print(XD_uParallelCarbon_outerdiv.min())
```
