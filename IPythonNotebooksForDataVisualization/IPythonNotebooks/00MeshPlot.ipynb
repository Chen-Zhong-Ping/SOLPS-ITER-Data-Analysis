```python
from packages.data_classes.mesh import Mesh
```

```python
import numpy
```

```python
import matplotlib.pyplot as plt

fig, ax_1 = plt.subplots()

from matplotlib.collections import LineCollection
```

```python
SD0203_mesh = Mesh('./Data/Geometry/SD/S20190203@25-6-66_24-6.v001.sno')
FS0308_mesh = Mesh('./Data/Geometry/FS/S20190308@25-10-50_25-6.v001.sno')
XD0130_mesh = Mesh('./Data/Geometry/XD/X20190130@31-12-68_26-6.v001.sno')
```

#### generate line collections

```python
SD_PoloidalSurfaces_SOL = LineCollection(SD0203_mesh.GridPoint_IyIx[SD0203_mesh.iycut : ])
SD_PoloidalSurfaces_PFR = LineCollection([SD0203_mesh.GridPoint_IyIx[i][:SD0203_mesh.ixcut[0]]+SD0203_mesh.GridPoint_IyIx[i][SD0203_mesh.ixcut[1]:] for i in range(0, SD0203_mesh.iycut)])
SD_PoloidalSurfaces_COR = LineCollection([SD0203_mesh.GridPoint_IyIx[i][SD0203_mesh.ixcut[0] : SD0203_mesh.ixcut[1]] + [SD0203_mesh.GridPoint_IyIx[i][SD0203_mesh.ixcut[0]]] for i in range(0, SD0203_mesh.iycut)])

SD_RadialPlanes = LineCollection(SD0203_mesh.GridPoint_IxIy)


FS_PoloidalSurfaces_SOL = LineCollection(FS0308_mesh.GridPoint_IyIx[FS0308_mesh.iycut : ])
FS_PoloidalSurfaces_PFR = LineCollection([FS0308_mesh.GridPoint_IyIx[i][:FS0308_mesh.ixcut[0]]+FS0308_mesh.GridPoint_IyIx[i][FS0308_mesh.ixcut[1]:] for i in range(0, FS0308_mesh.iycut)])
FS_PoloidalSurfaces_COR = LineCollection([FS0308_mesh.GridPoint_IyIx[i][FS0308_mesh.ixcut[0] : FS0308_mesh.ixcut[1]] + [FS0308_mesh.GridPoint_IyIx[i][FS0308_mesh.ixcut[0]]] for i in range(0, FS0308_mesh.iycut)])

FS_RadialPlanes = LineCollection(FS0308_mesh.GridPoint_IxIy)


XD_PoloidalSurfaces_SOL = LineCollection(XD0130_mesh.GridPoint_IyIx[XD0130_mesh.iycut : ])
XD_PoloidalSurfaces_PFR = LineCollection([XD0130_mesh.GridPoint_IyIx[i][:XD0130_mesh.ixcut[0]]+XD0130_mesh.GridPoint_IyIx[i][XD0130_mesh.ixcut[1]:] for i in range(0, XD0130_mesh.iycut)])
XD_PoloidalSurfaces_COR = LineCollection([XD0130_mesh.GridPoint_IyIx[i][XD0130_mesh.ixcut[0] : XD0130_mesh.ixcut[1]] + [XD0130_mesh.GridPoint_IyIx[i][XD0130_mesh.ixcut[0]]] for i in range(0, XD0130_mesh.iycut)])

XD_RadialPlanes = LineCollection(XD0130_mesh.GridPoint_IxIy)
```

set line colors

```python
SD_SurfaceColor_SOL = [(1, 0, 0, 1)] + [(0, 0, 0, 1)]*(SD0203_mesh.ny + 1 - SD0203_mesh.iycut)    # red, black
SD_SurfaceColor_PFR = [(0, 0, 0, 1)]
SD_SurfaceColor_COR = [(0, 0, 0, 1)]

SD_PoloidalSurfaces_SOL.set_color(SD_SurfaceColor_SOL)
SD_PoloidalSurfaces_PFR.set_color(SD_SurfaceColor_PFR)
SD_PoloidalSurfaces_COR.set_color(SD_SurfaceColor_COR)

SD_RadialPlanes.set_color((0, 0, 0, 1))


FS_SurfaceColor_SOL = [(1, 0, 0, 1)] + [(0, 0, 0, 1)]*(FS0308_mesh.ny + 1 - FS0308_mesh.iycut)    # red, black
FS_SurfaceColor_PFR = [(0, 0, 0, 1)]
FS_SurfaceColor_COR = [(0, 0, 0, 1)]

FS_PoloidalSurfaces_SOL.set_color(FS_SurfaceColor_SOL)
FS_PoloidalSurfaces_PFR.set_color(FS_SurfaceColor_PFR)
FS_PoloidalSurfaces_COR.set_color(FS_SurfaceColor_COR)

FS_RadialPlanes.set_color((0, 0, 0, 1))


XD_SurfaceColor_SOL = [(1, 0, 0, 1)] + [(0, 0, 0, 1)]*(XD0130_mesh.ny + 1 - XD0130_mesh.iycut)    # red, black
XD_SurfaceColor_PFR = [(0, 0, 0, 1)]
XD_SurfaceColor_COR = [(0, 0, 0, 1)]

XD_PoloidalSurfaces_SOL.set_color(XD_SurfaceColor_SOL)
XD_PoloidalSurfaces_PFR.set_color(XD_SurfaceColor_PFR)
XD_PoloidalSurfaces_COR.set_color(XD_SurfaceColor_COR)

XD_RadialPlanes.set_color((0, 0, 0, 1))
```

set line width

```python
width = 0.05

SD_PoloidalSurfaces_SOL.set_linewidth(width)
SD_PoloidalSurfaces_PFR.set_linewidth(width)
SD_PoloidalSurfaces_COR.set_linewidth(width)
SD_RadialPlanes.set_linewidth(width)

FS_PoloidalSurfaces_SOL.set_linewidth(width)
FS_PoloidalSurfaces_PFR.set_linewidth(width)
FS_PoloidalSurfaces_COR.set_linewidth(width)
FS_RadialPlanes.set_linewidth(width)

XD_PoloidalSurfaces_SOL.set_linewidth(width)
XD_PoloidalSurfaces_PFR.set_linewidth(width)
XD_PoloidalSurfaces_COR.set_linewidth(width)
XD_RadialPlanes.set_linewidth(width)
```
