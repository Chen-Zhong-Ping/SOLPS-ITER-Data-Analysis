some preparation calculation

```python
# first calculate total carbon ion density for later use
SD_nCarbonIon = SD_nIon3.IyIx + SD_nIon4.IyIx + SD_nIon5.IyIx + SD_nIon6.IyIx + SD_nIon7.IyIx + SD_nIon8.IyIx

FS_nCarbonIon = FS_nIon3.IyIx + FS_nIon4.IyIx + FS_nIon5.IyIx + FS_nIon6.IyIx + FS_nIon7.IyIx + FS_nIon8.IyIx

XD_nCarbonIon = XD_nIon3.IyIx + XD_nIon4.IyIx + XD_nIon5.IyIx + XD_nIon6.IyIx + XD_nIon7.IyIx + XD_nIon8.IyIx
```

1. calculate carbon poloidal velocity and its projection onto the field line

```python
# sum up charge states to get the total carbon ion flux
SD_FCarbonIon = SD_FIon3.IyIx + SD_FIon4.IyIx + SD_FIon5.IyIx + SD_FIon6.IyIx + SD_FIon7.IyIx + SD_FIon8.IyIx

FS_FCarbonIon = FS_FIon3.IyIx + FS_FIon4.IyIx + FS_FIon5.IyIx + FS_FIon6.IyIx + FS_FIon7.IyIx + FS_FIon8.IyIx

XD_FCarbonIon = XD_FIon3.IyIx + XD_FIon4.IyIx + XD_FIon5.IyIx + XD_FIon6.IyIx + XD_FIon7.IyIx + XD_FIon8.IyIx


# then calculate the flux density using cell geometry
SD_fCarbonIon = SD_FCarbonIon / (2 * numpy.pi * SD_cr.IyIx * SD_hy1.IyIx)

FS_fCarbonIon = FS_FCarbonIon / (2 * numpy.pi * FS_cr.IyIx * FS_hy1.IyIx)

XD_fCarbonIon = XD_FCarbonIon / (2 * numpy.pi * XD_cr.IyIx * XD_hy1.IyIx)


# then calculate the average carbon ion polodial velocity
SD_uPolCarbon = SD_fCarbonIon/SD_nCarbonIon

FS_uPolCarbon = FS_fCarbonIon/FS_nCarbonIon

XD_uPolCarbon = XD_fCarbonIon/XD_nCarbonIon


# then project the carbon ion polodial velocity to the direction along the field line using magnetic field components
SD_uPolCarbon_project = SD_uPolCarbon * SD_bb.IyIx / SD_bx.IyIx

FS_uPolCarbon_project = FS_uPolCarbon * FS_bb.IyIx / FS_bx.IyIx

XD_uPolCarbon_project = XD_uPolCarbon * XD_bb.IyIx / XD_bx.IyIx
```

2. calculate Carbon ion parallel velocity

```python
SD_uParallelCarbon = (SD_uIon3.IyIx * SD_nIon3.IyIx + SD_uIon4.IyIx * SD_nIon4.IyIx + SD_uIon5.IyIx * SD_nIon5.IyIx
                    + SD_uIon6.IyIx * SD_nIon6.IyIx + SD_uIon7.IyIx * SD_nIon7.IyIx + SD_uIon8.IyIx * SD_nIon8.IyIx)/SD_nCarbonIon

FS_uParallelCarbon = (FS_uIon3.IyIx * FS_nIon3.IyIx + FS_uIon4.IyIx * FS_nIon4.IyIx + FS_uIon5.IyIx * FS_nIon5.IyIx
                    + FS_uIon6.IyIx * FS_nIon6.IyIx + FS_uIon7.IyIx * FS_nIon7.IyIx + FS_uIon8.IyIx * FS_nIon8.IyIx)/FS_nCarbonIon

XD_uParallelCarbon = (XD_uIon3.IyIx * XD_nIon3.IyIx + XD_uIon4.IyIx * XD_nIon4.IyIx + XD_uIon5.IyIx * XD_nIon5.IyIx
                    + XD_uIon6.IyIx * XD_nIon6.IyIx + XD_uIon7.IyIx * XD_nIon7.IyIx + XD_uIon8.IyIx * XD_nIon8.IyIx)/XD_nCarbonIon
```

3. calculate effective non Coronal diffusive velocity along the field line 

```python
SD_uNCDParallelCarbon = SD_uPolCarbon_project - SD_uParallelCarbon

FS_uNCDParallelCarbon = FS_uPolCarbon_project - FS_uParallelCarbon

XD_uNCDParallelCarbon = XD_uPolCarbon_project - XD_uParallelCarbon
```

```python
# take out the outer divertor part

SD_uNCDParallelCarbon_outerdiv = numpy.transpose(SD_uNCDParallelCarbon[:, SD0203_mesh.ixcut[1]:])

FS_uNCDParallelCarbon_outerdiv = numpy.transpose(FS_uNCDParallelCarbon[:, FS0308_mesh.ixcut[1]:])

XD_uNCDParallelCarbon_outerdiv = numpy.transpose(XD_uNCDParallelCarbon[:, XD0130_mesh.ixcut[1]:])
```
