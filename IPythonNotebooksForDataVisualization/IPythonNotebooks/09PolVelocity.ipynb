```python
# import cell geometry data for calculation

# cell width in meters
SD_hy1 = Data2D(SD0203_mesh, './Data/SDP6/2D_data/geometry/hy1_2D')

FS_hy1 = Data2D(FS0308_mesh, './Data/FSP6/2D_data/geometry/hy1_2D')

XD_hy1 = Data2D(XD0130_mesh, './Data/XDP6/2D_data/geometry/hy1_2D')


# cell center radial coordinates in meters
SD_cr = Data2D(SD0203_mesh, './Data/SDP6/2D_data/geometry/cr_2D')

FS_cr = Data2D(FS0308_mesh, './Data/FSP6/2D_data/geometry/cr_2D')

XD_cr = Data2D(XD0130_mesh, './Data/XDP6/2D_data/geometry/cr_2D')
```

some preparation calculation

calculate carbon poloidal velocity

```python
# first calculate total carbon ion density for later use
SD_nCarbonIon = SD_nIon3.IyIx + SD_nIon4.IyIx + SD_nIon5.IyIx + SD_nIon6.IyIx + SD_nIon7.IyIx + SD_nIon8.IyIx

FS_nCarbonIon = FS_nIon3.IyIx + FS_nIon4.IyIx + FS_nIon5.IyIx + FS_nIon6.IyIx + FS_nIon7.IyIx + FS_nIon8.IyIx

XD_nCarbonIon = XD_nIon3.IyIx + XD_nIon4.IyIx + XD_nIon5.IyIx + XD_nIon6.IyIx + XD_nIon7.IyIx + XD_nIon8.IyIx


# then sum up charge states to get the total carbon ion flux
SD_FCarbonIon = SD_FIon3.IyIx + SD_FIon4.IyIx + SD_FIon5.IyIx + SD_FIon6.IyIx + SD_FIon7.IyIx + SD_FIon8.IyIx

FS_FCarbonIon = FS_FIon3.IyIx + FS_FIon4.IyIx + FS_FIon5.IyIx + FS_FIon6.IyIx + FS_FIon7.IyIx + FS_FIon8.IyIx

XD_FCarbonIon = XD_FIon3.IyIx + XD_FIon4.IyIx + XD_FIon5.IyIx + XD_FIon6.IyIx + XD_FIon7.IyIx + XD_FIon8.IyIx


# then calculate the flux density
SD_fCarbonIon = SD_FCarbonIon / (2 * numpy.pi * SD_cr.IyIx * SD_hy1.IyIx)

FS_fCarbonIon = FS_FCarbonIon / (2 * numpy.pi * FS_cr.IyIx * FS_hy1.IyIx)

XD_fCarbonIon = XD_FCarbonIon / (2 * numpy.pi * XD_cr.IyIx * XD_hy1.IyIx)


# then calculate the average carbon ion polodial velocity
SD_uPolCarbon = SD_fCarbonIon/SD_nCarbonIon

FS_uPolCarbon = FS_fCarbonIon/FS_nCarbonIon

XD_uPolCarbon = XD_fCarbonIon/XD_nCarbonIon


# take out the outer divertor part

SD_uPolCarbon_outerdiv = numpy.transpose(SD_uPolCarbon[:, SD0203_mesh.ixcut[1]:])

FS_uPolCarbon_outerdiv = numpy.transpose(FS_uPolCarbon[:, FS0308_mesh.ixcut[1]:])

XD_uPolCarbon_outerdiv = numpy.transpose(XD_uPolCarbon[:, XD0130_mesh.ixcut[1]:])
```

compare carbon velocity

```python
print(SD_uPolCarbon_outerdiv.max())
print(FS_uPolCarbon_outerdiv.max())
print(XD_uPolCarbon_outerdiv.max())

print(SD_uPolCarbon_outerdiv.min())
print(FS_uPolCarbon_outerdiv.min())
print(XD_uPolCarbon_outerdiv.min())
```
